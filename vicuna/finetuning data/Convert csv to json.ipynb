```python
# read csv file into pandas dataframe
import pandas as pd
df = pd.read_csv('/Users/svenschnydrig/Library/CloudStorage/OneDrive-SharedLibraries-UniversitätSt.Gallen/STUD-NLP Group Project - General/02 Group project/data/generated-data/df.csv')
```

```python
import json

# Initialize the list of conversations
conversations = []

# Loop through each row in the dataframe
for index, row in df.iterrows():
    # Create a new conversation object
    conversation = {
        "id": f"identity_{index}",
        "conversations": [
            {
                "from": "human",
                "value": row["Frage"]
            },
            {
                "from": "gpt",
                "value": row["Antwort"]
            }
        ]
    }
    # Append the new conversation to the list
    conversations.append(conversation)

# Write the conversations to a JSON file
with open('df.json', 'w') as f:
    json.dump(conversations, f)
```
